```python
import pandas as pd
import os

# 1. ç§»é™¤è£…é¥°å™¨ï¼Œä½¿å…¶æˆä¸ºæ™®é€šå‡½æ•°ï¼Œæ–¹ä¾¿åœ¨ Notebook ä¸­ç›´æŽ¥è°ƒç”¨
def mock_raw_data():
    return [
        {'hardModeLevel': 11, 'class': 'Priest', 'spec': 'Shadow', 'duration': 987942},
        {'hardModeLevel': 12, 'class': 'Mage', 'spec': 'Frost', 'duration': 1200000},
        {'hardModeLevel': 6, 'class': 'Warrior', 'spec': 'Fury', 'duration': 600000}
    ]

# 2. æµ‹è¯•é€»è¾‘å‡½æ•°
def test_data_cleaning_and_filtering(data):
    df = pd.DataFrame(data)
    df = df.rename(columns={'hardModeLevel': 'level', 'class': 'class_name', 'spec': 'spec_name'})
    df_filtered = df[(df['level'] >= 7) & (df['level'] <= 14)].copy()
    
    assert len(df_filtered) == 2, f"âŒ è¿‡æ»¤é”™è¯¯ï¼Œé¢„æœŸ2æ¡ï¼Œå®žé™…{len(df_filtered)}æ¡"
    print("âœ… æ•°æ®è¿‡æ»¤æµ‹è¯•é€šè¿‡ï¼")

def test_duration_conversion(data):
    df = pd.DataFrame(data)
    df['duration_min'] = df['duration'] / 60000
    assert round(df['duration_min'].iloc[0], 2) == 16.47
    print("âœ… æ—¶é—´è½¬æ¢æµ‹è¯•é€šè¿‡ï¼")

# 3. æ‰§è¡Œæµ‹è¯•
if __name__ == "__main__":
    print("ðŸš€ å¼€å§‹è¿è¡Œæ‰‹åŠ¨éªŒè¯...")
    raw_data = mock_raw_data() # æ­£å¸¸è°ƒç”¨
    test_data_cleaning_and_filtering(raw_data)
    test_duration_conversion(raw_data)
    print("ðŸŽ‰ é€»è¾‘éªŒè¯å…¨éƒ¨å®Œæˆï¼")
```

```python
import pandas as pd

# 1. ä»ä½ ä¹‹å‰ä¿å­˜çš„è·¯å¾„è¯»å–æ•°æ®
# è¯·ç¡®ä¿è·¯å¾„å’Œæ–‡ä»¶åä¸ä½ ä¿å­˜æ—¶ä¸€è‡´
df_final = pd.read_csv(r'..\data\raw\mplus_S3_7_14_COMPLETE.csv') 

# 2. ç°åœ¨å†è¿è¡Œè½¬æ¢é€»è¾‘å°±ä¸ä¼šæŠ¥é”™äº†
df_final['duration_min'] = (df_final['duration'] / 60000).round(2)
print("âœ… æ•°æ®è¯»å–å¹¶è½¬æ¢æˆåŠŸï¼")
print(df_final.head())
```

```python
# 1. ç¡®è®¤å½“å‰çš„åˆ—åï¼ˆç¡®ä¿ level, class_name, spec_name éƒ½å·²ç»å­˜åœ¨ï¼‰
print("å½“å‰åˆ—å:", df_final.columns)

# 1. æ¯«ç§’è½¬åˆ†é’Ÿï¼ˆå¦‚æœä¹‹å‰æ²¡è½¬çš„è¯ï¼‰
df_final['duration_min'] = (df_final['duration'] / 60000).round(2)

# 2. é‡æ–°å®šä¹‰æ ¸å¿ƒåˆ†æè¡¨ï¼Œè¿™æ¬¡å¸¦ä¸Š score å’Œ dungeon_name
# æ³¨æ„ï¼šç¡®ä¿ä½ çš„åˆ—åä¸ CSV ä¸­çš„åŸå§‹åç§°å®Œå…¨åŒ¹é…
analysis_df = df_final[[
    'class_name', 
    'spec_name', 
    'level', 
    'score',          # ğŸ‘ˆ ä½ è¦çš„è¯„åˆ†
    'duration_min', 
    'dungeon_name',   # ğŸ‘ˆ å‰¯æœ¬åï¼Œæ–¹ä¾¿çœ‹å“ªä¸ªèŒä¸šåœ¨å“ªä¸ªæœ¬å¼º
    'medal'           # ğŸ‘ˆ æ˜¯å¦é™æ—¶
]].copy()

# 3. ç®€å•æ¸…æ´—ï¼šå»æ‰é‚£äº› score å¼‚å¸¸çš„æ•°æ®ï¼ˆæ¯”å¦‚ 0 åˆ†ï¼‰
analysis_df = analysis_df[analysis_df['score'] > 0]

print("âœ… åˆ†æè¡¨å·²å°±ç»ªï¼Œå·²åŒ…å«è¯„åˆ†æ•°æ®ã€‚")
print(analysis_df.head())
```

```python
import os

# 1. å®šä¹‰å­˜å‚¨è·¯å¾„
output_dir = r'..\data\processed'
output_file = os.path.join(output_dir, 'mplus_S3_7_14_CLEANED.csv')

# 2. æ£€æŸ¥æ–‡ä»¶å¤¹æ˜¯å¦å­˜åœ¨ï¼Œä¸å­˜åœ¨å°±åˆ›å»ºä¸€ä¸ª
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 3. ä¿å­˜æ•°æ®ï¼ˆå»ºè®®ä¸ä¿å­˜ç´¢å¼• ignore_index=Trueï¼‰
analysis_df.to_csv(output_file, index=False)
print(f"âœ… æ¸…æ´—åçš„æ•°æ®å·²å­˜å…¥: {output_file}")
```

## Data Collection â€“ Warcraft Logs v2 API
This notebook handles authentication and raw data retrieval.

```python
from dotenv import load_dotenv
import os
import requests

load_dotenv()

CLIENT_ID = os.getenv("WCL_CLIENT_ID")
CLIENT_SECRET = os.getenv("WCL_CLIENT_SECRET")
assert CLIENT_ID and CLIENT_SECRET, "Missing WCL credentials. Check your .env file."

token_url = "https://www.warcraftlogs.com/oauth/token"
r = requests.post(
    token_url,
    data = {"grant_type": "client_credentials"},
    auth = (CLIENT_ID, CLIENT_SECRET),
    timeout = 30
    )
r.raise_for_status()

access_token = r.json()["access_token"]
print("Access token OK:", access_token[:12] + "...")
```

```python
endpoint = "https://www.warcraftlogs.com/api/v2/client"

query = """
query {
    worldData {
        expansions{ id name}
    }
}
"""

resp = requests.post(
    endpoint,
    json={"query":query},
    headers = {"Authorization": f"Bearer {access_token}"},
    timeout = 30
)
resp.raise_for_status()

data = resp.json()
print("GraphQL OK:", "data" in data and "worldData" in data["data"])
```

```python
# è·å–æ‰€æœ‰ Zone çš„ IDï¼Œç”¨äºå®šä½å¤§ç§˜å¢ƒèµ›å­£
zone_query = """
query {
  worldData {
    zones {
      id
      name
      frozen # å¦‚æœèµ›å­£å·²ç»“æŸï¼Œé€šå¸¸ä¸º true
    }
  }
}
"""
# è¿è¡Œæ–¹å¼åŒä½ æˆªå›¾ä¸­çš„é€»è¾‘ï¼Œæ‰¾åˆ°ç±»ä¼¼ "Mythic+ Season X" çš„ ID
# å‘é€è¯·æ±‚
resp = requests.post(
    endpoint,
    json={"query": zone_query},
    headers={"Authorization": f"Bearer {access_token}"},
    timeout=30
)

# data = resp.json()
# data
# # ä½ ä¼šçœ‹åˆ° dict_keys(['data'])


# æ£€æŸ¥çŠ¶æ€
if resp.status_code == 200:
    zones_data = resp.json()['data']['worldData']['zones']
    # æ‰“å°å‰ 10 ä¸ªçœ‹çœ‹
    for zone in zones_data:
        # æˆ‘ä»¬è¦æ‰¾åŒ…å« "Mythic+" å­—æ ·çš„ Zone
        if "Mythic+" in zone['name']:
            print(f"ID: {zone['id']} | Name: {zone['name']} | Frozen: {zone['frozen']}")
else:
    print(f"è¯·æ±‚å¤±è´¥ï¼ŒçŠ¶æ€ç ï¼š{resp.status_code}")
```

```python
import pandas as pd
import os

# 1. è®¾ç½®ç›®æ ‡å‰¯æœ¬ ID (ä»ä½ åˆšæ‰è¿è¡Œçš„ç»“æœä¸­é€‰ä¸€ä¸ª)
# 62660: Ara-Kara, City of Echoes
TARGET_EID = 62660 

# 2. ä¿®æ­£åçš„æ’åæŸ¥è¯¢è¯­å¥
# åœ¨ WCL v2 ä¸­ï¼Œå¤§ç§˜å¢ƒæ’åé€šå¸¸é€šè¿‡ encounter ä¸‹çš„ characterRankings è·å–
ranking_query = """
query ($eid: Int) {
  worldData {
    encounter(id: $eid) {
      name
      characterRankings(bracket: 10) # 10 ä»£è¡¨å¤§ç§˜å¢ƒ
    }
  }
}
"""

# 3. å‘é€è¯·æ±‚
resp = requests.post(
    endpoint,
    json={"query": ranking_query, "variables": {"eid": TARGET_EID}},
    headers={"Authorization": f"Bearer {access_token}"},
    timeout=30
)

res_json = resp.json()

# 4. æå–æ•°æ®å¹¶ä¿å­˜
if "data" in res_json and res_json['data']['worldData']['encounter']:
    # è·å–æ’ååˆ—è¡¨
    rankings = res_json['data']['worldData']['encounter']['characterRankings'].get('rankings', [])
    

# å…ˆè·å–ç¬¬ä¸€ä¸ªæ ·æœ¬ï¼Œçœ‹çœ‹å®ƒçš„å­—æ®µåˆ°åº•å«ä»€ä¹ˆ
if rankings:
    sample = rankings[0]
    print("API è¿”å›çš„å•æ¡è®°å½•å­—æ®µåï¼š", sample.keys())
    # æ‰“å°å‰ 3 æ¡æ•°æ®çš„æ ·æœ¬ï¼Œæ–¹ä¾¿æˆ‘ä»¬è§‚å¯Ÿæ•°æ®é•¿ä»€ä¹ˆæ ·
    for i in range(min(3, len(rankings))):
        print(f"æ ·æœ¬ {i}: {rankings[i]}")
```

```python
import pandas as pd
import os

# 1. æå–æ•°æ®
rankings = res_json['data']['worldData']['encounter']['characterRankings'].get('rankings', [])

if rankings:
    # 2. è½¬æ¢ä¸º DataFrame
    df_raw = pd.DataFrame(rankings)
    
    # 3. å­—æ®µæ˜ å°„ä¸é‡å‘½å (å°†æ¢æµ‹åˆ°çš„ hardModeLevel ç»Ÿä¸€ä¸ºæˆ‘ä»¬æƒ³è¦çš„ level)
    # åŒæ—¶æå–æˆ‘ä»¬åˆ†æéœ€è¦çš„å…³é”®åˆ—
    df_clean = df_raw.rename(columns={
        'hardModeLevel': 'level',
        'class': 'class_name',
        'spec': 'spec_name'
    })
    
    # 4. è¿‡æ»¤ 7-14 å±‚
    df_target = df_clean[(df_clean['level'] >= 7) & (df_clean['level'] <= 14)].copy()
    
    # 5. è®¡ç®—ä¸€äº›åŸºç¡€æŒ‡æ ‡ï¼ˆä¸ºé˜¶æ®µ 02 åšå‡†å¤‡ï¼‰
    # æ¯”å¦‚æŠŠæ¯«ç§’çº§çš„ duration è½¬ä¸ºåˆ†é’Ÿï¼Œæ–¹ä¾¿è§‚å¯Ÿ
    df_target['duration_min'] = df_target['duration'] / 60000
    
    print(f"âœ… æˆåŠŸï¼åœ¨å‰¯æœ¬ä¸­è·å–åˆ° {len(df_target)} æ¡ 7-14 å±‚æ•°æ®ã€‚")
    display(df_target[['name', 'class_name', 'spec_name', 'level', 'score', 'duration_min']].head())
    
    # 6. æŒä¹…åŒ–å­˜å‚¨
    os.makedirs('../data/raw', exist_ok=True)
    save_path = f"../data/raw/mplus_S3_EID_{TARGET_EID}_7_14.csv"
    df_target.to_csv(save_path, index=False)
    print(f"ğŸ’¾ æ•°æ®å·²å®‰å…¨å­˜æ¡£è‡³: {save_path}")
else:
    print("âš ï¸ æœªå‘ç°æ’è¡Œæ¦œæ•°æ®ã€‚")
```

```python
import time

# ä½ ä¹‹å‰æŸ¥åˆ°çš„ 8 ä¸ªå‰¯æœ¬ ID åˆ—è¡¨
mplus_v2_encounters = [62660, 12830, 62287, 62773, 62649, 112442, 112441, 62662]
all_dungeon_data = []

print("ğŸš€ å¼€å§‹å…¨å‰¯æœ¬æ•°æ®é‡‡é›†...")

for eid in mplus_v2_encounters:
    # æ„é€ è¯·æ±‚
    resp = requests.post(
        endpoint,
        json={"query": ranking_query, "variables": {"eid": eid}},
        headers={"Authorization": f"Bearer {access_token}"},
        timeout=30
    )
    
    res_json = resp.json()
    rankings = res_json['data']['worldData']['encounter']['characterRankings'].get('rankings', [])
    
    if rankings:
        df_temp = pd.DataFrame(rankings)
        # ç»Ÿä¸€å­—æ®µåå¹¶è¿‡æ»¤
        df_temp = df_temp.rename(columns={'hardModeLevel': 'level', 'class': 'class_name', 'spec': 'spec_name'})
        df_filtered = df_temp[(df_temp['level'] >= 7) & (df_temp['level'] <= 14)].copy()
        
        # è®°å½•å‰¯æœ¬åç§°
        df_filtered['dungeon_name'] = res_json['data']['worldData']['encounter']['name']
        all_dungeon_data.append(df_filtered)
        print(f"âœ… å·²å®Œæˆ: {df_filtered['dungeon_name']} | è·å–æ¡æ•°: {len(df_filtered)}")
    
    # ç¨å¾®ç­‰ä¸€ä¸‹ï¼Œå°Šé‡ API é™åˆ¶
    time.sleep(0.5)

# åˆå¹¶æ‰€æœ‰æ•°æ®
df_final_all = pd.concat(all_dungeon_data, ignore_index=True)
print(f"\nğŸ“Š é‡‡é›†ç»“æŸï¼æ€»è®¡è·å¾— {len(df_final_all)} æ¡èŒä¸šåˆ†ææ ·æœ¬ã€‚")

# ä¿å­˜æœ€ç»ˆæ€»è¡¨
df_final_all.to_csv('../data/raw/mplus_S3_ALL_DUNGEONS_7_14.csv', index=False)
```

```python
import time
import os
import pandas as pd

# 1. è®¾ç½®ç›®æ ‡å±‚æ•°åŒºé—´
target_levels = range(7, 15) # å³ 7, 8, ..., 14
mplus_v2_encounters = [62660, 62773, 62287, 62649, 12830, 62662, 112441, 112442]
all_data_frames = []

print("ğŸš€ å¼€å§‹æ·±åº¦æŠ“å– (7-14å±‚å…¨å‰¯æœ¬)...")

for eid in mplus_v2_encounters:
    for level in target_levels:
        # åœ¨ query ä¸­åŠ å…¥ bracket å‚æ•°æ¥é”å®šå±‚æ•°
        # æ³¨æ„ï¼šè¿™é‡Œæˆ‘ä»¬ä½¿ç”¨äº†å˜é‡ $lvl
        deep_query = """
        query ($eid: Int, $lvl: Int) {
          worldData {
            encounter(id: $eid) {
              name
              characterRankings(bracket: $lvl) 
            }
          }
        }
        """
        
        resp = requests.post(
            endpoint,
            json={"query": deep_query, "variables": {"eid": eid, "lvl": level}},
            headers={"Authorization": f"Bearer {access_token}"},
            timeout=30
        )
        
        res_json = resp.json()
        rankings = res_json['data']['worldData']['encounter']['characterRankings'].get('rankings', [])
        dungeon_name = res_json['data']['worldData']['encounter']['name']
        
        if rankings:
            df_temp = pd.DataFrame(rankings)
            # ç»Ÿä¸€å­—æ®µå
            df_temp = df_temp.rename(columns={'hardModeLevel': 'level', 'class': 'class_name', 'spec': 'spec_name'})
            df_temp['dungeon_name'] = dungeon_name
            
            all_data_frames.append(df_temp)
            print(f"âœ… {dungeon_name} | ç¬¬ {level} å±‚ | æŠ“å–åˆ° {len(df_temp)} æ¡")
        
        # é¢‘ç‡æ§åˆ¶ï¼Œé˜²æ­¢è¢« API å±è”½
        time.sleep(0.3)

# åˆå¹¶å¹¶å­˜æ¡£
if all_data_frames:
    df_final = pd.concat(all_data_frames, ignore_index=True)
    os.makedirs('../data/raw', exist_ok=True)
    df_final.to_csv('../data/raw/mplus_S3_7_14_COMPLETE.csv', index=False)
    print(f"\nğŸ“Š æ·±åº¦é‡‡é›†å®Œæˆï¼æ€»æ ·æœ¬é‡: {len(df_final)}")
```

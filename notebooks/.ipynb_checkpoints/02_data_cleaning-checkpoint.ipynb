```python
import pandas as pd

# 假设 df_raw 是你从 01 脚本中拿到的原始数据
# rankings 列里面通常是一个包含 5 个字典的列表 [{}, {}, {}, {}, {}]

def clean_mplus_data(df_raw):
    # 1. 炸开嵌套列表：将一行(队)变五行(人)
    df_exploded = df_raw.explode('rankings').reset_index(drop=True)
    
    # 2. 提取每个人员的详细信息
    # 注意：WCL 返回的字段名可能包含 classID, specID, bracketData (即装等)
    player_info = df_exploded['rankings'].apply(pd.Series)
    
    # 3. 合并回主表（保留层数、限时时间等环境数据）
    df_final = pd.concat([df_exploded.drop(columns=['rankings']), player_info], axis=1)
    
    # 4. 这里的 'bracketData' 通常是装等，转为数值型
    df_final['ilvl'] = pd.to_numeric(df_final['bracketData'], errors='coerce')
    
    return df_final

# 应用逻辑
df_players = clean_mplus_data(df_filtered)
```
